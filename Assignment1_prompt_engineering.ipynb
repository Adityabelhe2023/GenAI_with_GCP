```python
from google import genai
import os
from IPython.display import display, Markdown

# Configure API key (replace with your actual key or use environment variable)
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY", "YOUR_API_KEY_HERE")

# Initialize the client
client = genai.Client(api_key=GOOGLE_API_KEY)

# Set the model to use
MODEL_ID = "gemini-2.5-flash"  # You can choose other models as needed

print("✅ Gemini API configured successfully!")
```

```python
prompt = """You are an AI assistant that classifies customer feedback above into sentiment categories: Positive, Neutral, or Negative. 
For each feedback entry:
1. Read the text carefully.
2. Think step by step about the tone used.
3. Explain your reasoning briefly.
4. Output the final sentiment label along with the review that was provided.
"""
chain_of_thoughts_shot_prompt = """Feedback: "The delivery was super fast and the product quality exceeded my expectations!"
Reasoning: Mentions fast delivery and exceeded expectations → strong satisfaction.
Sentiment: Positive

Feedback: "The product works fine, but the packaging was just average."
Reasoning: Product is acceptable, packaging is average → balanced, no strong emotion.
Sentiment: Neutral

Feedback: "I’m really disappointed. The item stopped working after two days."
Reasoning: Clear dissatisfaction and product failure → negative sentiment.
Sentiment: Negative

Feedback: "Customer support was helpful, but I had to wait a long time before reaching them."
Reasoning: Helpful support (positive) but long wait (negative) → mixed, overall neutral.
Sentiment: Neutral

Feedback: "I love this brand! Every purchase feels worth the money."
Reasoning: Strong positive emotion and satisfaction with value → positive sentiment.
Sentiment: Positive
"""
```

```python
my_query="My product flow was halted in between becasue of connection issue. "
```

```python
final_prompt=my_query+prompt+chain_of_thoughts_shot_prompt
print(f"prompt passed {final_prompt}")
```

```python
response = client.models.generate_content(
    model=MODEL_ID,
    contents=final_prompt

)
print("chain_of_thoughts Prompt:")
print (f"{final_prompt}")
print(response.text)
```

# Gemini API Parameters -

This notebook demonstrates all the key parameters you can use with Gemini API

## Setup: Import and Initialize

````python
import os
from google import genai
from dotenv import load_dotenv

# Load API key
load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")

# Initialize client
client = genai.Client(api_key=api_key)
MODEL = "gemini-2.5-flash"

print("✅ Gemini client ready!")
print(f"📌 Using model: {MODEL}")
````

## 1️⃣ Temperature - Controls Randomness

**Range:** 0.0 to 2.0
- **0.0** = Deterministic, focused, predictable
- **1.0** = Balanced creativity
- **2.0** = Maximum creativity, randomness

**Use cases:**
- Low (0.0-0.3): Facts, code, structured data
- Medium (0.7-1.0): General conversation, content writing
- High (1.5-2.0): Creative writing, brainstorming

````python
# Same prompt, different temperatures
prompt = "Write a one-sentence description of a cat."

print("=" * 80)
print("🧊 LOW Temperature (0.0) - Predictable, Focused")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 0.0}
)
print(response.text)

print("\n" + "=" * 80)
print("🌡️ MEDIUM Temperature (1.0) - Balanced")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 1.0}
)
print(response.text)

print("\n" + "=" * 80)
print("🔥 HIGH Temperature (1.8) - Creative, Random")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 1.8}
)
print(response.text)

print("\n💡 Notice how the responses vary in creativity!")
````

## 2️⃣ Max Output Tokens - Controls Response Length

**What it does:** Limits how many tokens (roughly words) the model can generate

**Typical values:**
- **50-100**: Very short, concise answers
- **500-800**: Medium paragraphs
- **2000+**: Long, detailed responses

**Note:** 1 token ≈ 0.75 words (varies by language)

````python
prompt = "Explain what Python programming language is."

print("=" * 80)
print("📏 SHORT Response (max_output_tokens=50)")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"max_output_tokens": 50, "temperature": 0.3}
)
print(response.text)
print(f"\n✂️ Response was cut short due to token limit")

print("\n" + "=" * 80)
print("📜 MEDIUM Response (max_output_tokens=200)")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"max_output_tokens": 200, "temperature": 0.3}
)
print(response.text)

print("\n" + "=" * 80)
print("📚 LONG Response (max_output_tokens=500)")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"max_output_tokens": 500, "temperature": 0.3}
)
print(response.text)
````

## 3️⃣ Top-P (Nucleus Sampling) - Controls Token Selection

**Range:** 0.0 to 1.0

**What it does:** The model considers only the most probable tokens whose cumulative probability reaches P

**Examples:**
- **0.1**: Very narrow selection (only top 10% probable tokens)
- **0.9**: Broader selection (considers 90% of probability mass)
- **1.0**: Considers all possible tokens

**Common setting:** 0.9 (balanced)

````python
prompt = "Complete this sentence: The future of AI will"

print("=" * 80)
print("🎯 LOW top_p (0.1) - Very Focused")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 0.8, "top_p": 0.1, "max_output_tokens": 100}
)
print(response.text)

print("\n" + "=" * 80)
print("🌐 HIGH top_p (0.95) - More Diverse")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 0.8, "top_p": 0.95, "max_output_tokens": 100}
)
print(response.text)

print("\n💡 Higher top_p allows more creative word choices!")
````

## 4️⃣ Top-K - Limits Token Choices

**What it does:** Only consider the K most likely tokens at each step

**Examples:**
- **top_k=1**: Always pick the most likely token (deterministic)
- **top_k=10**: Choose from top 10 most likely tokens
- **top_k=40**: Choose from top 40 tokens (default)
- **top_k=100**: Broader selection

**Use case:** Control vocabulary diversity

````python
prompt = "List 5 creative names for a coffee shop."

print("=" * 80)
print("🔒 LOW top_k (10) - Limited Choices")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 0.9, "top_k": 10, "max_output_tokens": 150}
)
print(response.text)

print("\n" + "=" * 80)
print("🔓 HIGH top_k (100) - More Options")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={"temperature": 0.9, "top_k": 100, "max_output_tokens": 150}
)
print(response.text)

print("\n💡 Higher top_k can produce more unique word combinations!")
````

## 5️⃣ Combining Parameters - Real-World Scenarios

Here are recommended parameter combinations for different use cases:

````python
prompt = "Explain quantum computing."

print("" + "=" * 80)
print("📊 SCENARIO 1: Factual/Technical Writing")
print("Settings: Low temp, moderate tokens")
print("=" * 80)

response = client.models.generate_content(
    model=MODEL,
    contents=prompt,
    config={
        "temperature": 0.2,
        "max_output_tokens": 200,
        "top_p": 0.8,
        "top_k": 40
    }
)
print(response.text)

print("\n" + "=" * 80)
print("🎨 SCENARIO 2: Creative Writing")
print("Settings: High temp, high tokens")
print("=" * 80)

creative_prompt = "Write a creative opening line for a sci-fi story."
response = client.models.generate_content(
    model=MODEL,
    contents=creative_prompt,
    config={
        "temperature": 1.5,
        "max_output_tokens": 150,
        "top_p": 0.95,
        "top_k": 100
    }
)
print(response.text)

print("\n" + "=" * 80)
print("💬 SCENARIO 3: Conversational Assistant")
print("Settings: Balanced parameters")
print("=" * 80)

chat_prompt = "How can I improve my productivity?"
response = client.models.generate_content(
    model=MODEL,
    contents=chat_prompt,
    config={
        "temperature": 0.7,
        "max_output_tokens": 300,
        "top_p": 0.9,
        "top_k": 40
    }
)
print(response.text)
````

## 📚 Parameter Summary Table

| Parameter | Range | Effect | Use When |
|-----------|-------|--------|----------|
| **temperature** | 0.0 - 2.0 | Controls randomness | 0.2 for facts, 1.5 for creativity |
| **max_output_tokens** | 1 - 8192 | Limits response length | Set based on desired output size |
| **top_p** | 0.0 - 1.0 | Nucleus sampling | 0.9 is a good default |
| **top_k** | 1 - 100+ | Limits token choices | 40 is standard, higher = more diverse |

## 💡 Quick Tips

1. **Start with defaults** (temp=0.7, top_p=0.9, top_k=40) and adjust as needed
2. **Temperature is the most important** parameter for controlling output style
3. **Lower temperature** = More consistent results when running multiple times
4. **Higher temperature** = Different results each time (good for brainstorming)
5. **Don't set both top_p and top_k too low** - it restricts creativity too much
6. **max_output_tokens** helps control costs and response time
